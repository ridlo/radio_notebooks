```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fringe(uv, x, y):
    res = 0.0
    for i,u in enumerate(uv): 
        res += np.cos(2*np.pi*(uv[i][0]*x/206265 + uv[i][1]*y/206265)) # in arcsec
        
    return res
```

### 2 antenna

```python
uv = [[150000, 300000]]
```

```python
n = 256
x = np.linspace(-5, 5, n) 
y = np.linspace(-5, 5, n)
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, fringe(uv, X, Y), 8, alpha=.75, cmap='hot')
```
